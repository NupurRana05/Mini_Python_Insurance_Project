# Mini Python  Insurance Project  

```python
import pandas as pd

data = {
    'age': [25, 45, 52, 37, 62],
    'gender': ['Female', 'Male', 'Male', 'Female', 'Female'],
    'bmi': [23.4, 27.8, 31.2, 22.9, 26.1],
    'children': [0, 2, 1, 3, 0],
    'smoker': ['No', 'Yes', 'No', 'Yes', 'No'],
    'region': ['southeast', 'southwest', 'northeast', 'northwest', 'southeast'],
    'charges': [3200, 16800, 9000, 14000, 3240]
}

insurance_df = pd.DataFrame(data)
insurance_df

```

```python
# Basic Info
print(insurance_df.info())
```

```python
# Discriptive statistics
print(insurance_df.describe())
```

```python
# check for null values
print(insurance_df.isnull().sum().sum())
```

```python
# value counts for categorical variables
print("Gender Distribution")
print(insurance_df['gender'].value_counts())
print("\n Smoker Distribution")
print(insurance_df['smoker'].value_counts())

```

```python
# Display Correlation
print("\n Correlation matrix:")
print(insurance_df.corr(numeric_only=True))
```

```python
# python script to clean  the data
```

```python
# Drop Duplicates
insurance_df =insurance_df.dropna()
```

```python
# handle possible missing values
```

```python
insurance_df = insurance_df.fillna({
    'age': insurance_df['age'].median(),
    'bmi': insurance_df['bmi'].mean(),
    'charges': insurance_df['charges'].mean(),
    'children': insurance_df['children'].mean(),
    'gender': insurance_df['gender'].mode()[0],
    'smoker': insurance_df['smoker'].mode(),
    'region': insurance_df['region'].mode()[0]
})

# standardize categorical values
insurance_df['gender'] = insurance_df['gender'].str.title()
insurance_df['smoker'] = insurance_df['smoker'].str.title()
insurance_df['region'] = insurance_df['region'].str.lower()

display(insurance_df)
    
```

```python
# Import Libraries Create KPI Charts
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# set plot style
sns.set(style="whitegrid")
# Average charges by smoking status
plt.figure(figsize=(6,4))
sns.barplot(data=insurance_df, x='smoker', y='charges', ci=None)
plt.title('Average  Insurance Charges by Smoking Status')
plt.tight_layout()
plt.show()

# Distribution of charges
plt.figure(figsize=(6,4))
sns.histplot(insurance_df['charges'], bins=10, kde=True)
plt.title('Distribution of Onsurance charges')
plt.tight_layout()
plt.show()
```

```python
# Changes by region
plt.figure(figsize=(6,4))
sns.boxplot(data=insurance_df, x='region', y='charges')
plt.title('Insurance Charges by Region')
plt.tight_layout()
plt.show()
```

```python
# correlation BMI VS Changes
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot( data=insurance_df, x='bmi',  y='charges', hue='smoker'
)
plt.title("BMI VS CHARGES DISTRIBUTION")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.pagesizes import A4

```

```python
# PDF OF AGE
plt.figure()
plt.hist(insurance_df['age'], bins=5, density=True)
plt.xlabel("Age")
plt.ylabel("Probability Density")
plt.title("Probability Distribution of Age")
age_pdf = "age_pdf.png"
plt.savefig(age_pdf)
plt.show()
```

```python
# PDF OF BMI
plt.figure()
plt.hist(insurance_df['bmi'], bins=5, density=True)
plt.xlabel("BMI")
plt.ylabel("Probability Density")
plt.title("Probability Distribution of BMI")
bmi_pdf = "bmi_pdf.png"
plt.savefig(bmi_pdf)
plt.show()
```

```python
# PDF OF INSURANCE CHARGES
plt.figure()
plt.hist(insurance_df['charges'], bins=5, density=True)
plt.xlabel("Charges")
plt.ylabel("Probability Density")
plt.title("Probability Distribution of Insurance Charges")
charges_pdf = "charges_pdf.png"
plt.savefig(charges_pdf)
plt.show()
```

```python
# Export PDF Report
pdf_report = "Probability_Distribution_Report.pdf"

doc = SimpleDocTemplate(pdf_report, pagesize=A4)
styles = getSampleStyleSheet()

story = []

story.append(Paragraph("<b>Probability Distribution Function (PDF) Report</b>", styles['Title']))
story.append(Spacer(1, 12))

story.append(Paragraph(
    "This report presents Probability Distribution Functions (PDFs) "
    "for numerical features in the insurance dataset including Age, BMI, "
    "and Insurance Charges.",
    styles['Normal']
))

story.append(Spacer(1, 20))
story.append(Image(age_pdf, width=400, height=300))
story.append(Spacer(1, 20))
story.append(Image(bmi_pdf, width=400, height=300))
story.append(Spacer(1, 20))
story.append(Image(charges_pdf, width=400, height=300))

doc.build(story)

print("Probability Distribution Function (PDF) Report created successfully!")
```
